## EDA

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
from utils import buildDataset, dictLetras
```

`dictLetrasInv` contiene los mismos elementos que dictLetras, pero con las llaves y los valores invertidos

```python
dictLetrasInv = {v: k for k, v in dictLetras.items()}
```

 Se cargan y preparan un conjunto de datos de imágenes para su posterior uso en tareas de aprendizaje automático, donde `X` representará las características de las imágenes y `y` las etiquetas correspondientes.

```python
height, width = 200, 200

X, y = buildDataset("../Data/Raw Data", dictLetras, width, height)
```

Cambiar la forma del arreglo X de modo que las imágenes se representen como tensores tridimensionales, donde cada imagen tiene dimensiones de altura, ancho y tres canales (escala RGB).

```python
X = X.reshape(-1, height, width, 3)
```

Este dataset contiene 6300 imágenes, en este caso de 200x200 pixeles y a color, divididas en 21 categorías correspondientes a las letras que se representan de forma estática, donde cada categoría tiene 300 imágenes.

```python
X.shape, y.shape
```

El resultado del gráfico de barras nos permite visualizar la distribución de clases del conjunto de datos. Todas las barras tienen la misma altura (cada letra tiene 300 imágenes), eso indica que las clases están balanceadas, lo que es importante para evitar sesgos hacia clases mayoritarias.

Se muestran las letras que se utilizarán, las que no se incluyen son las señas que son móviles: j, k, ñ, q, x y z; ya que para estas se neceistaría registro de los movimientos en vídeo. de cada letra contamos con 300 imagenes que nos dan un total de 6300.

```python
shapeTag = {dictLetrasInv[ID]:count for ID, count in zip(*np.unique(y, return_counts=True))}
plt.bar(shapeTag.keys(), shapeTag.values())
plt.show()
```

Ahora podemos visualizar la primera imagen en el conjunto de datos X. Esto es útil para verificar cómo se ve una imagen específica en el conjunto de datos.

En este caso podemos vizualizar la letra R.

```python
plt.figure()
plt.axis('off')
plt.title('Letra: ' + dictLetrasInv[y[0]] +  ' ; Shape: ' + str(X[0].shape))
plt.imshow(cv.cvtColor(X[0], cv.COLOR_BGR2RGB))
plt.grid(False)
plt.show()
```
